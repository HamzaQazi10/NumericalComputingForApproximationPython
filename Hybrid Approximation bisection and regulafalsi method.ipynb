```python
import math
def cal(x):
    ans=((math.sin(x))-(5*x)+2)
    return ans


rootA=0.1
rootB=1.4
def BiRegRoot(rootA,rootB,ErrMax):
    cond=0
    k=0
    ErrMax=ErrMax
    while(cond==0 and k<100):
        #print("--------------Bisection-----------------")
        mid=(rootA+rootB)/2
        print("Root Point is "+str(mid))
        funA=cal(rootA)
        funB=cal(rootB)
        funMid=cal(mid)
        print("Function of A is "+str(funA))
        print("Function of B is "+str(funB))
        print("Function of Mid is "+str(funMid))
        if((funA*funMid)<0):
            rootB=mid
        elif((funB*funMid)<0):
            rootA=mid
        else:        
            print("The Approximate root is "+str(mid))
            cond=1
        print("A is "+str(rootA))
        print("B is "+str(rootB))
        k=k+1
        print("K is "+str(k))
        midCheck=(rootA+rootB)/2
        print("Tolerance "+str(abs(cal(midCheck))))
        if(abs(cal(midCheck)) > ErrMax):
            cond=0
        else:
            cond=1
            break
        #print("--------------Regula Falsi-----------------")

        funA=cal(rootA)
        funB=cal(rootB)
        mid=((rootA*funB)-(rootB*funA))/(funB-funA)
        print("Mid is "+str(mid))
        funMid=cal(mid)
        print("Function of A is "+str(funA))
        print("Function of B is "+str(funB))
        print("Function of Mid is "+str(funMid))
        if((funA*funMid)<0):
            rootB=mid
        elif((funB*funMid)<0):
            rootA=mid
        else:
            print("The Approximate root is "+str(mid))
            cond=1
        print("A is "+str(rootA))
        print("B is "+str(rootB))
        k=k+1
        print("K is "+str(k))
        print("Tolerance "+str(abs(cal(midCheck))))
        if(abs(cal(midCheck)) > ErrMax):
            cond=0
        else:
            cond=1
            break
        print("---------------------------------------------------------------------------------------------")
    return midCheck
BiRegRoot(rootA,rootB,0.00001)
```
