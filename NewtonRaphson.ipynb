```python
#Hamza Arshad Qazi
#SP19-BCS-084

import math

print("-----------NewtonRaphson METHOD-------------")
def cal(x):
    ans=ans=((x**2)-x-2)
    return ans
def cal1(x):
    ans=((2*x)-1)
    return ans

k=0
rootA=1 # You may take the input from the user here in rootA variable
#rootB=3
cond=0
funMidPrev=0  # To store Previous Approximation
errorCond=0
while(cond==0 and k<15):
    funA=cal(rootA)
    funN=cal1(rootA)
    mid=(rootA-(funA/funN))
    print("Root Point is "+str(mid))
    
    print("Function of A is "+str(funA))
    print("Function of Derivative of A is "+str(funN))
    check=rootA
    rootA=mid
    
    print("A is "+str(rootA))
    print("Derivative of A is "+str(rootA))
    k=k+1
    print("Number of Iterations are: "+str(k))
    if(funA==0):
        print("The Final Root Is: "+str(rootA))
        break
    if(errorCond!=0):# Check for Previous Value if available or not
        print("Present Approximation"+str(funA))
        print("Previous Approximation"+str(funMidPrev))
        RelativeApproximateError=(funA-funMidPrev)/funA
        print("Error is "+str(RelativeApproximateError*100)+str('%'))
    else:
        print("No Previous Approximation, So unable to Calculate Relative Approximate Error ")
    errorCond=1
    funMidPrev=funA
    if(rootA==check):
        print("------------------The Root is "+str(rootA)+"------------------")
        break
    print("---------------------------------Next Iteration--------------------------------------------")         
```
